#pandas
Bibliothèque spécialisée dans la manipulation et l’analyse des données tabulaires via les DataFrames. Elle offre des outils puissants pour le nettoyage, la sélection, la transformation, l’agrégation et l’exploration statistique des données. Pandas est particulièrement adaptée pour gérer des données hétérogènes et manquantes, et pour préparer les données avant analyse ou modélisation
#numpy
Cœur du calcul numérique en Python, offrant des structures de données efficaces (tableaux multidimensionnels ndarrays) et un large éventail de fonctions mathématiques et statistiques bas niveau. NumPy est optimisé pour des calculs vectoriels rapides et sert de base à d’autres bibliothèques comme Pandas ou scikit-learn. Il permet de calculer rapidement des statistiques descriptives, des opérations sur matrices, et des analyses numériques.
#matplotlib
Bibliothèque de visualisation 2D et 3D, très flexible, qui permet de créer tous types de graphiques statistiques classiques (histogrammes, courbes, scatter plots, boxplots). Utilisée pour comprendre les distributions, visualiser des corrélations et interpréter des résultats. Matplotlib est la base sur laquelle Seaborn s’appuie pour fournir des visualisations plus accessibles.
#seaborn
Extension de Matplotlib orientée vers la visualisation statistique avancée. Elle fournit des fonctions de haut niveau pour tracer facilement des graphiques comme les heatmaps, plots de distribution, pair plots, graphiques de régression, facilitant l’analyse exploratoire des données. Seaborn est conçu pour fonctionner de façon fluide avec les DataFrames Pandas, rendant la visualisation plus intuitive et esthétique.
#scikit-learn
Bibliothèque incontournable pour le machine learning classique. Construite sur NumPy, SciPy et Matplotlib, elle propose une API uniforme et simple pour appliquer de nombreux algorithmes de classification, régression, clustering, réduction de dimensionnalité, ainsi que des outils pour le pré-traitement et l’évaluation statistique des modèles. Elle est optimisée pour la performance grâce à des implémentations en Cython/Fortran, et s’intègre bien avec Pandas pour la préparation des données et avec Matplotlib/Seaborn pour la visualisation des résultats.
#jupyter
Environnement interactif qui permet de combiner code, visualisation et documentation dans un même document dynamique. L’environnement Jupyter est largement utilisé pour expérimenter, analyser et communiquer la démarche statistique, facilitant ainsi la reproductibilité des analyses avec ces bibliothèques.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Charger un dataset en ligne (exemple : ventes d'une entreprise)
url = "https://raw.githubusercontent.com/datasets/gdp/master/data/gdp.csv"
df = pd.read_csv(url)
```

fonction "info"
Une fonction dans un code est un ensemble d’instructions regroupées sous un nom, qui s’exécutent lorsqu’on l’appelle. Elle peut recevoir des paramètres pour traiter des données et retourner un résultat. Les fonctions permettent de structurer le code, de le rendre réutilisable et plus clair.

```python
# Informations sur le dataset
print(df.info())
```

```python
# Statistiques descriptives
print(df.describe())
```

```python
# Statistiques pour une colonne spécifique
colonne = df['Value']  # Adapter selon vos données

print(f"Moyenne: {colonne.mean():.2f}")
print(f"Médiane: {colonne.median():.2f}")
print(f"Mode: {colonne.mode()[0]:.2f}")
print(f"Écart-type: {colonne.std():.2f}")
print(f"Variance: {colonne.var():.2f}")
print(f"Min: {colonne.min():.2f}")
print(f"Max: {colonne.max():.2f}")
```

data visualisation

```python
# Configuration du style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Créer une figure avec plusieurs sous-graphiques
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Histogramme
axes[0, 0].hist(colonne, bins=30, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Distribution des valeurs', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Valeur')
axes[0, 0].set_ylabel('Fréquence')

# Ajouter la moyenne et la médiane
axes[0, 0].axvline(colonne.mean(), color='red', linestyle='--',
                   label=f'Moyenne: {colonne.mean():.2f}')
axes[0, 0].axvline(colonne.median(), color='green', linestyle='--',
                   label=f'Médiane: {colonne.median():.2f}')
axes[0, 0].legend()

# 2. Boxplot
axes[0, 1].boxplot(colonne.dropna(), vert=True)
axes[0, 1].set_title('Boxplot - Détection des outliers', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Valeur')

# 3. Densité
colonne.plot(kind='density', ax=axes[1, 0])
axes[1, 0].set_title('Courbe de densité', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Valeur')

# 4. QQ-plot (pour tester la normalité)
from scipy import stats
stats.probplot(colonne.dropna(), dist="norm", plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot (Test de normalité)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
